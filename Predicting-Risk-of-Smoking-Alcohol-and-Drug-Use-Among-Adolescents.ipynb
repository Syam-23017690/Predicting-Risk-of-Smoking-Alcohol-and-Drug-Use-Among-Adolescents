```python
#loading SchoolData
# Path to your uploaded ZIP file
Zipfile_path2 = 'C:/Users/Syam7/OneDrive/My project/New folder/school-pupils-and-their-characteristics_2023-24.zip'

# Open the ZIP file
with zipfile.ZipFile(Zipfile_path2) as z:
    # List all files inside
    file_list = z.namelist()
    print("Files in ZIP:", file_list)
    

# Dictionary to store loaded DataFrames
dfs = {}

# Common encodings to try
encodings = ['utf-8', 'cp1252', 'latin1', 'ISO-8859-1']

with zipfile.ZipFile(Zipfile_path2, 'r') as z:
    for file in z.namelist():
        if file.endswith('.csv'):
            for enc in encodings:
                try:
                    with z.open(file) as f:
                        dfs[file] = pd.read_csv(f, encoding=enc)
                    print(f"Loaded {file} with encoding {enc}, shape: {dfs[file].shape}")
                    break  # Stop trying encodings once successful
                except UnicodeDecodeError:
                    continue
            else:
                print(f"Failed to load {file} due to encoding issues.")
```
